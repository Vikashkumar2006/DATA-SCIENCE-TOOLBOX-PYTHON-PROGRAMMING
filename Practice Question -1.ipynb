```python
import numpy as np
import random
import scipy.stats as stats
a=[random.gauss(50,20) for x in range(30)]
b=[random.gauss(55,15) for x in range(30)]
print(a)
print(b)
t_stat, p_value=stats.ttest_1samp(a,45,axis=0)
print("T-statistics: ",t_stat)
print("P-value:",p_value)
print(np.mean(a))
```
