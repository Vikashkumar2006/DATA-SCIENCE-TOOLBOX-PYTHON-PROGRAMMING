```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("WineQT.csv")
```

```python
df.head()
```

```python
hi=df.loc[df["alcohol"].idxmax()][["alcohol","quality"]]
hi
```

```python
wi=df.loc[(df["alcohol"]>=10) & (df["alcohol"]<=12) & (df["quality"]>=6)]
wi["Id"].count()
```

```python
avg=df[df["quality"]>=7].mean()
avg
```

```python
den_avg=df["density"].mean()
Alc_avg=df["alcohol"].mean()
den=df[(df["density"]>den_avg) & (df["alcohol"]<Alc_avg)]
den
```

```python
s_wine=df[(df['pH']>3.5) & (df['citric acid']>0.4) & (df["quality"]>=7)]
s_wine
```

```python
df["sulphur_ratio"]=df["free sulfur dioxide"]/df["total sulfur dioxide"]
avg_sul=df.groupby("quality")["sulphur_ratio"].mean()
avg_sul
```
