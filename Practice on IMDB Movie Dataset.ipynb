```python
# For correlation and covariance
import pandas as pd
```

```python
df=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
df.isnull().sum()
```

```python
df=df[['Revenue (Millions)','Metascore','Rating']].dropna()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
correlation_matrix=df.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.show()
```

```python
#Close to +1 → Strong positive linear relationship

#Close to -1 → Strong negative linear relationship

#Close to 0 → No linear relationship

covaranice_matric=df.cov()
covaranice_matric
```
