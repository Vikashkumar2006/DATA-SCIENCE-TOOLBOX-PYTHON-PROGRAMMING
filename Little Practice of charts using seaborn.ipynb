```python
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('weight-height.csv')
```

```python
df.head(10)
```

```python
sns.boxplot(df['Height'])
```

```python
# Percentile method
upper_limit =df['Height'].quantile(0.99)
lower_limit=df['Height'].quantile(0.01)
print(upper_limit,lower_limit)
```

```python
#using Trimming method
newdf=df.loc[(df['Height']>=lower_limit) & (df['Height']<=upper_limit)]
sns.boxplot(newdf['Height'])
```

```python
#implementing covariance Height vs. weight
covariance=df['Height'].cov(df['Weight'])
print(covariance)
```

```python
sns.scatterplot(x=df['Height'],y=df['Weight'],hue=df['Gender'],alpha=0.7)
```
