```python
df.fillna({'County':df['County'].mode()[0]},inplace=True)
df.fillna({'City':df['City'].mode()[0]},inplace=True)
df.fillna({'Postal Code':df['Postal Code'].mode()[0]},inplace=True)
df.fillna({'Electric Range':df['Electric Range'].mean()},inplace=True)
df.fillna({'Base MSRP':df['Base MSRP'].mean(),'Legislative District':df['Legislative District'].mean(),'Vehicle Location':df['Vehicle Location'].mode()[0],'Electric Utility':df['Electric Utility'].mode()[0],'2020 Census Tract':df['2020 Census Tract'].mean()},inplace=True)
```

```python
#Summary Statistics

print(df.describe())
print(df.describe(include='object'))
```

```python
#Distribution of Model Years
plt.figure(figsize=(12, 6))
sns.countplot(x="Model Year", data=df, order=sorted(df["Model Year"].unique()), hue="Model Year", palette="coolwarm", legend=False)
plt.xticks(rotation=45)
plt.title("Distribution of Model Years")
plt.xlabel("Model Year")
plt.ylabel("Count")
plt.show()
```

```python
#Top vehicle Makes & Models
top_makes = df["Make"].value_counts().nlargest(10)

sns.barplot(x=top_makes.index, y=top_makes.values, hue=top_makes.index, palette="magma", legend=False)

plt.title("Top 10 Vehicle Makes")
plt.xlabel("Make")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Top vehicle Makes & Models
top_makes = df["Make"].value_counts().nlargest(10)

sns.barplot(x=top_makes.index, y=top_makes.values, hue=top_makes.index, palette="magma", legend=False)

plt.title("Top 10 Vehicle Makes")
plt.xlabel("Make")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Vehicle Models
plt.figure(figsize=(12,6))
top_models = df["Model"].value_counts().nlargest(10)

sns.barplot(x=top_models.index, y=top_models.values, hue=top_models.index, palette="viridis", legend=False)

plt.title("Top 10 Vehicle Models")
plt.xlabel("Model")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Electric Range Distribution

plt.figure(figsize=(12,6))
sns.histplot(df["Electric Range"].dropna(),bins=30,kde=True,color="green")
plt.title("Electric Range Distribution")
plt.xlabel("Electric Range (Miles)")
plt.ylabel("Count")
plt.show()
```

```python
#Electric Vehicle Type Breakdown
plt.figure(figsize=(8,6))
df["Electric Vehicle Type"].value_counts().plot(kind="pie",autopct="%1.1f%%",colors=["skyblue","orange"])
plt.title("Electric Vehicle Type Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df["Electric Vehicle Type"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.show()
```

```python
#Correlation Analysis
numeric_df=df.select_dtypes(include=["number"])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(),annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
numeric_df=df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Longitude", y="Latitude", data=df, alpha=0.5, color="blue")
plt.title("Electric Vehicle Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
#Time Trends in Electric Vehicle Adoption
plt.figure(figsize=(12, 6))
df["Model Year"].value_counts().sort_index().plot(kind="line", marker="o", color="purple")
plt.title("Electric Vehicle Adoption Over Time")
plt.xlabel("Model Year")
plt.ylabel("Number of Vehicles")
plt.grid(True)
plt.show()
```

```python
#Clean Alternative Fuel Vehicle (CAFV) Eligibility Analysis
plt.figure(figsize=(12,6))
df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].value_counts().plot(kind="bar", color="cyan")
plt.title("CAFV Eligibility Status")
plt.xlabel("Eligibility")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Base MSRP'], y=df['Electric Range'], alpha=0.5)
plt.xlabel("Base MSRP ($)")
plt.ylabel("Electric Range (Miles)")
plt.title("Electric Range vs. MSRP")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(y=df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'], hue=df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'], palette="pastel", legend=False)
plt.xlabel("Number of Vehicles")
plt.ylabel("CAFV Eligibility")
plt.title("CAFV Eligibility Distribution")
plt.show()
```

```python
#z-score test
upper_limit=df['Electric Range'].mean()+3*df['Electric Range'].std()
lower_limit=df['Electric Range'].mean()-3*df['Electric Range'].std()
outlinear= df[(df['Electric Range']>=upper_limit) | (df['Electric Range']<=lower_limit)]
print(len(outlinear))
```

```python
#outlinear removal using z test
newdf=df.loc[(df['Electric Range']<upper_limit) & (df['Electric Range']>lower_limit)]
sns.boxplot(newdf['Electric Range'])
print(len(newdf))
```
