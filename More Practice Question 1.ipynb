```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
product = ['Phone', 'Tablet', 'Laptop', 'Headphones', 'Smartwatch']
sold = [50, 30, 20, 100, 45]
unit = [500, 300, 1000, 150, 200]

data = []
for i in range(len(product)):
    total = sold[i] * unit[i]
    data.append({
        'Product Name': product[i],
        'Quantity': sold[i],
        'Unit': unit[i],
        'Total Sales': total
    })

df = pd.DataFrame(data)
print(df[df['Total Sales']>10000])
```

```python
df=pd.read_csv("mtcars.csv")
df.head()
```

```python
high_mpg = df[df['mpg'] > 25]
df.rename(columns={'Unnamed: 0':'car'},inplace=True)
print(df[df['mpg']>25][['car','mpg']])

powerful_inefficient = df[(df['hp'] > 150) & (df['mpg'] < 20)]
print(powerful_inefficient[['car', 'hp', 'mpg']])

top3_mpg = df.nlargest(3, 'mpg')[['car', 'mpg']]
print(top3_mpg)
```

```python
bills = [2300, 3100, 1800, 2900, None, 2750, 3250, 2400]
bills_array = np.array(bills, dtype=np.float64)
print("a) NumPy Array of Bills:\n", bills_array)


greater_than_2500 = bills_array[bills_array > 2500]
print("b) Bills greater than ₹2500:\n", greater_than_2500)


max_bill = np.nanmax(bills_array)
min_bill = np.nanmin(bills_array)
print("c) Max bill:", max_bill)
print("   Min bill:", min_bill)


total_sum = np.nansum(bills_array)
print("d) Total sum of bills (excluding None):", total_sum)


mean_bill = np.nanmean(bills_array)
print("e) Mean of bills (excluding None):", mean_bill)

```
