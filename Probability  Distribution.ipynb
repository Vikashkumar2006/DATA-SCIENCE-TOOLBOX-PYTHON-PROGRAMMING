```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fig,axes=plt.subplots(2,2,figsize=(14,10))

uni=np.random.uniform(low=0,high=10,size=1000)
sns.histplot(data=uni,kde=True,ax=axes[0,0],color='red')
axes[0,0].set_title('Uniform Distribution (size=1000)')


# Loc means mean, scale here means Standard Deviation
norm =np.random.normal(loc=0,scale=1,size=1000)
sns.histplot(norm,kde=True,bins=30,ax=axes[0,1],color='purple')
axes[0,1].set_title('Normal Distribution with mean 0 and S.D. 1')




#binomial
bino=np.random.binomial(n=100,p=0.5,size=1000)
sns.histplot(bino,kde=True,bins=13,ax=axes[1,0],color='salmon')
axes[1,0].set_title('Binomial Distribution with 100 trials, p=0.5')


#poisson Distribution ,lam means: Lambda value
pois=np.random.poisson(lam=4,size=1000)
sns.histplot(pois,kde=True,discrete=True,bins=12,ax=axes[1,1],color='red')
axes[1,1].set_title('Poisson Distribution with lambda value 4')

plt.show()
```
