```python
import pandas as pd
import os
c=os.getcwd()
print(c)
```

```python
cust =pd.read_csv("cust1.csv")
```

```python
cust.head()
```

```python
order=pd.read_csv("order1.csv")
```

```python
order.head()
```

```python
merge_csv=pd.merge(cust,order)
```

```python
merge_csv
```

```python
#using on=common column
joined_table=pd.merge(cust,order,on="CustId")
#if common column  is having different names in both tables
joined_table=pd.merge(cust,order,left_on="CustId",right_on="CustId")
print(joined_table)
```

```python
#joined on multiple columns
joined_table=pd.merge(cust,order,on=["CustId","OrderId"])
joined_table
```

```python
outer_join=pd.merge(cust,order,on="CustId",how="outer")
outer_join
```

```python
left_join=pd.merge(cust,order,on="CustId",how="left")
left_join
```

```python
right_join=pd.merge(cust,order,on="CustId",how="right")
right_join
```

```python
concat_result=pd.concat([left_join,right_join],axis=0)
#axis=0 for rows and axis =1 for columns
print(concat_result)
```
